```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

```python
# Data preparation
# Data preparation as x and y
```

```python
y = df['logS']
y
```

```python
x = df.drop('logS',axis=1)
x
```

```python
# Data Scplitting

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
# Model building
# Linear regression
```

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Model training
model.fit(x_train, y_train)
```

```python
# Apply the model to make predictions
```

```python
y_pred_train = model.predict(x_train)
# Evaluating the model
y_pred_test = model.predict(x_test)
```

```python
y_pred_train
```

```python
y_pred_test
```

```python
# Evaluate model performance
```

```python
from sklearn.metrics import mean_squared_error, r2_score
model_train_mse = mean_squared_error(y_train, y_pred_train)
model_train_r2 = r2_score(y_train, y_pred_train)
# Model classification 
model_test_mse = mean_squared_error(y_test, y_pred_test)
model_test_r2 = r2_score(y_test, y_pred_test)
```

```python
print("Model MSE (train): ", model_train_mse)
print("Model R2 (train): ", model_train_r2)
print("Model MSE (test): ", model_test_mse)
print("Model R2 (test): ", model_test_r2)
```

```python
print(model_train_mse,model_train_r2,model_test_mse,model_test_r2)
```

```python
model_result = pd.DataFrame(['LinearRegression', model_train_mse,model_train_r2, model_test_mse, model_test_r2]).transpose()

model_result
```

```python
# Renamae model columns
model_result.columns = ['Method','Training MSE','Traing R2','Test MSE','Test R2']
model_result
```

```python
# Random Forest
```

```python
# Traning the model
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(max_depth=2, random_state=100)
rf_model.fit(x_train, y_train)
```

```python
# Applying the model to make a prediction
y_pred_train_rf = rf_model.predict(x_train)
y_pred_test_rf = rf_model.predict(x_test)
```

```python
# Evalute model performance
from sklearn.metrics import mean_squared_error,r2_score
model_train_mse_rf = mean_squared_error(y_train, y_pred_train_rf)
model_train_r2_rf = r2_score(y_train, y_pred_train_rf)

model_test_mse_rf = mean_squared_error(y_test, y_pred_test_rf)
model_test_r2_rf = r2_score(y_test, y_pred_test_rf)
```

```python
rf_result = pd.DataFrame(['Random forest', model_train_mse_rf, model_train_r2_rf, model_test_mse_rf, model_test_r2_rf]).transpose()
rf_result.columns = ['Method','Training MSE','Traing R2','Test MSE','Test R2']
rf_result
```

```python
# Combine two table together
# Model comparison
df_models = pd.concat([model_result, rf_result])
df_models
```

```python
# Set an index into the model table
df_models.reset_index(drop=True)
```
