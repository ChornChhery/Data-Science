```python
import pandas as pd
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.rename(columns={'Annual Income (k$)':'AnnualIncome','Spending Score (1-100)':'SpendingScore'},inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df[['Age','AnnualIncome','SpendingScore']])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['AnnualIncome'],df['SpendingScore'],s=50)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Annual Income vs Spending Score')
plt.show()
```

```python
# Apply K-mean Algorithm
from sklearn.cluster import KMeans
x = df[['AnnualIncome','SpendingScore']]

wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++',max_iter=300, n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss)
plt.xlabel("Number of Cluster")
plt.ylabel("WCSS")
plt.title("Elbow method to determine optimal N of cluster")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++',max_iter=300, n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(x)
```

```python
df['Cluster'] = y_kmeans
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=y_kmeans,s=50,cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0 ],centers[:,1],c='red',s=200,alpha=0.75,marker='X')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Customer Segments')
plt.show()
```

```python
# Selecting features for clustering
x = df[['Age','AnnualIncome','SpendingScore']]

# Determining the optimal number of cluster using the elbow method
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++',max_iter=300, n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("Number of Cluster")
plt.ylabel("WCSS")
plt.title("Elbow Method to Determine Dptimal N of Cluster")
plt.show()
```

```python
# Applying K-means with the chosen number of clusters
kmeans = KMeans(n_clusters=6, init='k-means++',max_iter=300, n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(x)
df['ClusterAgeIncomeSpend'] = y_kmeans
```

```python
# Visualizing the clusters in a 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(df['Age'],df['AnnualIncome'],df['SpendingScore'],c=df['ClusterAgeIncomeSpend'],s=50,cmap='viridis')
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_xlabel('Spending Score')
plt.title('Customer Segments based on Age, Annual Income, and Spending Score')
plt.show()
```
