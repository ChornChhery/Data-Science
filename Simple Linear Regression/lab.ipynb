```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
url = "https://github.com/TheEconomist/big-mac-data/raw/master/output-data/big-mac-adjusted-index.csv"
df = pd.read_csv(url, parse_dates=['date'])
df.head()
```

```python
df.info()
```

```python
df = df[(df['date'].dt.year==2024) & (df['date'].dt.month==1)]
df.head()
```

```python
sns.lmplot(x='GDP_bigmac',y='dollar_price',data=df, ci=None)
# ci is confindence interval
# lm is linear Model
```

```python
sns.jointplot(x='GDP_bigmac',y='dollar_price',data=df, ci=None, kind='reg',color='orange')
```

```python
## Simple Linear Regression Model 
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
df['GDP_bigmac']

# The value of this is call Series
```

```python
df[['GDP_bigmac']]

## The value of this is call DataFrame
```

```python
X= df[['GDP_bigmac']] # expect 2D array not Series (1D)
y = df.dollar_price
```

```python
model.fit(X, y)
```

```python
model.score(X, y)  # R-squared of model
```

```python
model.intercept_
```

```python
model.coef_

# The value of this have only one because we use x as an dependent variable only one
```

```python
model.intercept_ + model.coef_*9000

# Method 1
# Formula => y = a + bx
# To predict that if that country have a GDP 9000$ what is the price of MC Donal that they should sell
```

```python
model.predict([[9000]])

# Method 2
# To predict that if that country have a GDP 9000$ what is the price of MC Donal that they should sell
```

```python
model.predict([[9000],[40000]])
```

```python
# Want to predict that if that country have a GDP from 5000 until 50000$ 
np.arange(5000,50001,2500)
```

```python
model.predict(np.arange(5000,50001,2500).reshape(-1,1)) # Array 2D

# The value of price depends on the GDP price
```

```python
np.linspace(5000,50000,20).reshape(-1,1)
```

```python
model.predict(np.linspace(5000,50000,20).reshape(-1,1))
```

```python
# Create a Series 
s = pd.Series(np.linspace(5000,50000,20)) # 1D Array
s
```

```python
model.predict(s.to_frame())
```

```python
dx = pd.DataFrame(np.linspace(5000,50000,20), columns=['gdp'])
dx
```

```python
model.predict(dx)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# formula: response ~ predictor + predictor
model_a = smf.ols(formula='dollar_price ~ GDP_bigmac',data=df).fit()
model_a.summary()
```

```python
print(model_a.summary())
```
