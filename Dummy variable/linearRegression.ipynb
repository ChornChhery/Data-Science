```python
import pandas as pd
```

```python
df = pd.read_csv("https://github.com/prasertcbs/tutorial/raw/master/msleep.csv")
df.head()
```

```python
df.info()
```

```python
## Check NA in 'vore' column
```

```python
df[df.vore.isna()]
```

```python
## Drop NA in 'vore' column
df.dropna(subset=['vore'], axis='index', inplace=True)
df
```

```python
## Create dummy variable
```

```python
dummies = pd.get_dummies(df['vore'])
dummies
```

```python

dm = pd.concat([df,dummies], axis="columns")
dm
```

```python
## Sklearn Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
dt = dm[['vore','bodywt','carni','herbi','omni','insecti','sleep_total']].dropna(axis='index') # is mean that Na column I want to delete
X = dt[['bodywt','insecti','herbi','omni']]
```

```python
model.fit(X, dt.sleep_total)
```

```python
model.score(X, dt.sleep_total) # R-squared 
```

```python
model.coef_
#Answer array of this: X = dt[['bodywt','insecti','herbi','omni']] 
# we got 4 answer of array of this
```

```python
## Statmodels: Linear regression
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df.columns
```

```python
# Formula : response ~ predictor + predictor . Noted: ols is Ordinary list square
model_a = smf.ols(formula='sleep_total ~ bodywt + awake + awake + sleep_cycle', data=df).fit()
```

```python
model_a.pvalues
```

```python
print(model_a.summary())
```
