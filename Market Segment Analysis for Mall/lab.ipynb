```python
# 12. Category by Gender and payment methods
# px.sunburst(df,path=['Gender','Category','Payment Method'],values='Purchase Amount (USD)',title='Category by gender and payment method')

# Aggregate data
grouped = df.groupby(['Gender', 'Category', 'Payment Method'])['Purchase Amount (USD)'].sum().reset_index()

# Pivot data to prepare for stacked bar chart
pivot_data = grouped.pivot_table(
    index=['Gender', 'Category'], 
    columns='Payment Method', 
    values='Purchase Amount (USD)', 
    fill_value=0
)

# Plot the stacked bar chart
pivot_data.plot(kind='bar', stacked=True, figsize=(12, 8), cmap='Paired')

# Add title and labels
plt.title('Category by Gender and Payment Method', fontsize=16)
plt.xlabel('Gender and Category', fontsize=12)
plt.ylabel('Purchase Amount (USD)', fontsize=12)
plt.legend(title='Payment Method', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Conclusion 12:
# _ Mostly the shopping is done by 'Male'
# _ Mostly the men/female shop for clothing followed by accessories
# _ The most preferred payment method for male is 'credit card' and for female it's 'Venmo'
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score,classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Encode categorical variables

category = ['Gender', 'Category', 'Location', 'Season', 'Subscription Status', 
                    'Payment Method', 'Shipping Type', 'Promo Code Used', 'Preferred Payment Method']
df_encoded = df.copy()
```

```python
label_encoders = {}
for col in category:
    label_encoders[col] = LabelEncoder()
    df_encoded[col] = label_encoders[col].fit_transform(df[col])
```

```python
# Map "Frequency of Purchases" to numeric
freq_numer = {
    'Daily': 30,
    'Weekly': 4,
    'Fortnightly': 2,
    'Monthly':1
}
df_encoded['Frequency of Purchases'] = df['Frequency of Purchases'].map(freq_numer)
```

```python
# Encode categorical columns
df_encoded = pd.get_dummies(
    df,
    columns=[
        "Gender",
        "Item Purchased",
        "Category",
        "Location",
        "Size",
        "Color",
        "Season",
        "Subscription Status",
        "Payment Method",
        "Shipping Type",
        "Discount Applied",
        "Promo Code Used",
        "Preferred Payment Method",
        "Frequency of Purchases"
    ],
    drop_first=True
)

# Target column and feature set
x_reg = df_encoded.drop(columns=["Purchase Amount (USD)"])
y_reg = df_encoded["Purchase Amount (USD)"]
```

```python
# # Machine Learning model
# # Step 1: Using Linear Regression to Predict 'Purchase Amount (USD)'

# x_reg = df_encoded.drop(['Purchase Amount (USD)','Item Purchased'],axis=1)
# y_reg = df_encoded['Purchase Amount (USD)']

# x_train_reg, x_test_reg, y_train_reg, y_test_reg = train_test_split(x_reg,y_reg, test_size=0.2,random_state=42)
```

```python
# train test split
x_train_reg,x_test_reg,y_train_reg,y_test_reg = train_test_split(x_reg,y_reg, test_size=0.2,random_state=42)

scaler = StandardScaler()
x_train_reg = scaler.fit_transform(x_train_reg)
x_test_reg = scaler.transform(x_test_reg)
```

```python
# Train Model
lin_reg = LinearRegression()
lin_reg.fit(x_train_reg, y_train_reg)
```

```python
# Prediction
y_pred = lin_reg.predict(x_test_reg)
y_pred
```

```python

print(f'Mean Squared Error: {mean_squared_error(y_test_reg,y_pred):.2f}')
```
